```python
import matplotlib.pyplot as plt
import numpy as np
fig1 = plt.figure() # графическое окно figure 1
ax1 = fig1.add_subplot(1, 1, 1) # добавляем график 1 на figure 1
plt.title('function $y = e^{x}$')
ax1.plot(x,y) # строим функцию на графике
fig1.show()
fig2 = plt.figure() # графическое окно figure 2
ax2 = fig2.add_subplot(1, 1, 1)
plt.title('function $y = ln(x)$')
ax2.plot(y,x) # поменяем местами x и y, построим обратную функцию x(y)
fig2.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
x = np.linspace(-np.pi,np.pi,10)
x1 = x
x2 = x
y = np.sin(x)
y1 = np.sin(x)-2
y2 = np.sin(x)+1
plt.plot(x,y) # строим функцию 1 на графике
plt.plot(x1,y1) # строим функцию 2 на графике
plt.plot(x2,y2) # строим функцию 3  на графике
plt.xlabel('x')
plt.ylabel('y')
plt.title('functions $y = sin(x), y = sin(x)-2 , y = sin(x)+1$') # название для графика 1
plt.plot(x,y,'-og',x1,y1,'-om',x2,y2,'-ok') #цвет графиков, тип линий, формы маркеров
plt.grid()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
x=np.linspace(-2.0*np.pi,2.0*np.pi,100)
handle1, = plt.plot(x,np.cos(x),label='y=cos(x)')
handle2, = plt.plot(x,np.cos(2.0*x),label='y=cos(2x)')
handle3, = plt.plot(x,np.cos(0.5*x),label='y=cos(x/2)')
plt.legend(handles=[handle1,handle2,handle3])
plt.axhline(y=0, color='k') # отобразим ось x
plt.axvline(x=0, color='k') # отобразим ось y
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
x=np.linspace(-5,5,100) 
plt.subplot(2,3,1)
y = np.abs(np.abs(x)-2)
plt.plot(x,y,'-r') # строим функцию 1 на графике
plt.grid(True) # включаем отображение координатной сетки
plt.subplot(2,3,2)
y1 = np.abs(np.abs(x-2)-2)
plt.plot(x,y1,'-g') # строим функцию 1 на графике
plt.grid(True) # включаем отображение координатной сетки
plt.subplot(2,3,3)
y2 = np.abs(np.abs(x+2)-2)
plt.plot(x,y2,'-k') # строим функцию 1 на графике
plt.grid(True) # включаем отображение координатной сетки
plt.subplot(2,3,4)
y3 = np.abs(np.abs(2*x)-2)
plt.plot(x,y3,'-y') # строим функцию 1 на графике
plt.grid(True) # включаем отображение координатной сетки
plt.subplot(2,3,5)
y4 = np.abs(np.abs(0.5*x)-2)
plt.plot(x,y4,'-b') # строим функцию 1 на графике
plt.grid(True) # включаем отображение координатной сетки
plt.subplot(2,3,6)
y5 = np.abs(np.abs(-x)-2)
plt.plot(x,y5,'-m') # строим функцию 1 на графике
plt.grid(True) # включаем отображение координатной сетки
```

```python
import matplotlib.pyplot as plt
import numpy as np
x=np.linspace(0,np.pi,10)
y1 = plt.plot(x,np.cos(x),':g')
y2 = plt.plot(x,1/(np.cos(x)),'-.r')
plt.grid(True) # включаем отображение координатной сетки
plt.xlabel('x')
plt.ylabel('y')
plt.axhline(y=0, color='k') # отобразим ось x
plt.axvline(x=0, color='k') # отобразим ось y
plt.axis('equal')
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
x = np.linspace(-2,2,100)
x1 = np.linspace(np.exp(-2),np.exp(2),100)
x2 = np.linspace(-2,np.exp(2),100)
y = np.exp(x)
y1 = np.exp(x1)
y2 = x2
plt.plot(x,y) # строим функцию 1 на графике
plt.plot(y1,x1) # строим функцию 2 на графике
plt.plot(x2,y2) # строим функцию 3  на графике
plt.xlabel('x')
plt.ylabel('y')
plt.title('functions $y = e^{x}, y = ln(x) , y = x$') # название для графика 1
plt.plot(x,y,'-g',y1,x1,'-m',x2,y2,'-y') #цвет графиков, тип линий, формы маркеров

plt.grid()
plt.show()
```
